```python
def generate_text(prompt):
  input_tokens = tokenizer.encode(prompt,return_tensors='pt')
  output_logits = model(input_tokens).logits[:,-1,:]
  output_tokens = torch.argmax(output_logits)
  output_text = tokenizer.decode(output_tokens)
  return output_text
```

```python
generate_text("hello")
```

```python
spc_map = tokenizer.special_tokens_map
```

```python
def generate_text_repetative(prompt):
  final_word = prompt
  output_text = ""
  while output_text != spc_map['eos_token']:
  # for i in range(20):
    input_tokens = tokenizer.encode(prompt,return_tensors='pt')
    output_logits = model(input_tokens).logits[:,-1,:]
    output_tokens = torch.argmax(output_logits)
    output_text = tokenizer.decode(output_tokens)
    final_word += output_text
    prompt = final_word
  return final_word

print(generate_text_repetative("Hello"))
```

```python
#using past key values:
def generate_with_kvcache(prompt, max_new_tokens=50):
    input_ids = tokenizer.encode(prompt, return_tensors='pt')

    # first forward
    outputs = model(input_ids)
    past = outputs.past_key_values

    generated = input_ids

    for _ in range(max_new_tokens):
        # only feed last token
        next_token = generated[:, -1:]

        outputs = model(next_token, past_key_values=past, use_cache=True)
        past = outputs.past_key_values

        logits = outputs.logits[:, -1, :]
        next_id = torch.argmax(logits, dim=-1)

        generated = torch.cat([generated, next_id.unsqueeze(0)], dim=-1)

        if next_id.item() == tokenizer.eos_token_id:
            break

    return tokenizer.decode(generated[0])
```

```python
generate_with_kvcache("I am a", max_new_tokens=50)
```
